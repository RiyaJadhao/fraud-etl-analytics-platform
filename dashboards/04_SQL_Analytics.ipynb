Use Catalog & Schema

```sql
CREATE CATALOG IF NOT EXISTS fraud_catalog;
CREATE SCHEMA IF NOT EXISTS fraud_catalog.fraud_schema;
USE fraud_catalog.fraud_schema;

```

Silver Layer Validation

```sql
SELECT COUNT(*) AS total_silver_records
FROM silver_transactions;

SELECT *
FROM silver_transactions
LIMIT 10;
```

Transaction Volume Analysis

```sql
SELECT
  type,
  COUNT(*) AS total_transactions,
  SUM(amount) AS total_amount
FROM silver_transactions
GROUP BY type
ORDER BY total_amount DESC;
```

Fraud Analysis by Type

```sql
SELECT
  type,
  COUNT(*) AS fraud_count,
  SUM(amount) AS fraud_amount
FROM silver_transactions
WHERE isFraud = 1
GROUP BY type
ORDER BY fraud_amount DESC;
```

Gold Layer Verification

```sql
SELECT *
FROM gold_transactions
ORDER BY fraud_amount DESC;
```

Fraud Trend Over Time

```sql
SELECT
  step,
  COUNT(*) AS fraud_count,
  SUM(amount) AS fraud_amount
FROM silver_transactions
WHERE isFraud = 1
GROUP BY step
ORDER BY step;
```

High-Risk Transactions

```sql
SELECT
  step,
  type,
  amount,
  nameOrig,
  nameDest
FROM silver_transactions
WHERE isFraud = 1
ORDER BY amount DESC
LIMIT 20;
```

Data Validation  Record Counts

```sql
-- Total transactions
SELECT COUNT(*) AS total_transactions
FROM silver_transactions;

-- Total fraud transactions
SELECT COUNT(*) AS total_fraud_transactions
FROM silver_transactions
WHERE isFraud = 1;
```

Transaction Distribution by Type

```sql
SELECT
    type,
    COUNT(*) AS total_transactions,
    SUM(amount) AS total_amount
FROM silver_transactions
GROUP BY type
ORDER BY total_amount DESC;
```

Fraud Analysis by Transaction Type (CORE QUERY)

```sql
SELECT
    type,
    COUNT(*) AS fraud_count,
    SUM(amount) AS fraud_amount
FROM silver_transactions
WHERE isFraud = 1
GROUP BY type
ORDER BY fraud_amount DESC;
```

Fraud Rate by Transaction Type

```sql
SELECT
    type,
    COUNT(*) AS total_transactions,
    SUM(isFraud) AS fraud_transactions,
    ROUND(SUM(isFraud) / COUNT(*) * 100, 2) AS fraud_rate_percentage
FROM silver_transactions
GROUP BY type
ORDER BY fraud_rate_percentage DESC;
```

High-Value Fraud Transactions

Top Destination Accounts Involved in Fraud

```sql
SELECT
    nameDest,
    COUNT(*) AS fraud_count,
    SUM(amount) AS fraud_amount
FROM silver_transactions
WHERE isFraud = 1
GROUP BY nameDest
ORDER BY fraud_amount DESC
LIMIT 10;
```

Flagged vs Actual Fraud Comparison

```sql
SELECT
    isFlaggedFraud,
    COUNT(*) AS transaction_count,
    SUM(isFraud) AS actual_fraud
FROM silver_transactions
GROUP BY isFlaggedFraud;
```

Fraud Percentage Contribution by Type

```sql
SELECT
    type,
    SUM(amount) AS fraud_amount,
    ROUND(
        SUM(amount) * 100 /
        (SELECT SUM(amount) FROM silver_transactions WHERE isFraud = 1),
        2
    ) AS fraud_percentage_contribution
FROM silver_transactions
WHERE isFraud = 1
GROUP BY type
ORDER BY fraud_percentage_contribution DESC;
```
